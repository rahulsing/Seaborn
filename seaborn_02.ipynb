Bivariant Visualization 

Relationship between two variable or data point

```python
import os 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine_data=pd.read_csv("./datasets/wine-reviews/winemag-data_first150k.csv")
```

```python
white_wind_data=pd.read_csv("./datasets/wine-reviews/winequality-white.csv",
                            names=["Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar","Chlorides",
                                   "Free Sulfur Dioxide","Total Sulfur Dioxide","Density","pH","Sulphates","Alcohol","Quality"],
                            skiprows=1,sep=r'\s*;\s*',engine='python')
```

```python
sns.jointplot(x="Free Sulfur Dioxide",y="Total Sulfur Dioxide",data=white_wind_data)

#plot a scatter plot 
#plots the distribution on the plot top and right as histogram
```

```python
sns.jointplot(x="Free Sulfur Dioxide",y="Total Sulfur Dioxide",data=white_wind_data,xlim={0,100},ylim={0,250})

# Resultant Zoomed-in plot by specifing the range of x and y range
```

```python
with sns.axes_style('white'):
    sns.jointplot(x="Free Sulfur Dioxide",y="Total Sulfur Dioxide",data=white_wind_data,
                  kind='hex',color='k',
                  xlim={0,100},ylim={0,250})

# Resultant: hexbin plot by specifiing the kind='hex', histogram with bivariant data
# Important : Scatter plot with hue variation, drak/light of the point represnt the height of historam at that point
```

```python
with sns.axes_style('white'):
    sns.jointplot(x="Free Sulfur Dioxide",y="Total Sulfur Dioxide",data=white_wind_data,
                  kind='kde',color='k',
                  xlim={0,100},ylim={0,250})
# kind='kde' , for the bivarinat create a KDE plot
```

```python
sns.kdeplot(white_wind_data["Free Sulfur Dioxide"],white_wind_data["Total Sulfur Dioxide"],clip=(0,250))

plt.xlim(0,100)
plt.ylim(0,250)

# Resultant: KDE using kdeplot with clip to limit range
```

```python
f,ax=plt.subplots(figsize=(8,5))

sns.kdeplot(white_wind_data["Free Sulfur Dioxide"],white_wind_data["Total Sulfur Dioxide"],clip=(0,250))

sns.rugplot(white_wind_data["Free Sulfur Dioxide"],color='g',height=0.05)
sns.rugplot(white_wind_data["Free Sulfur Dioxide"],vertical=True,height=0.05)

plt.xlim(-20,100)
plt.ylim(-50,400)
 
```
