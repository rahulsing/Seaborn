```python
import os 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
white_wine_data=pd.read_csv("./datasets/wine-reviews/winequality-white.csv",
                            names=["Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar","Chlorides",
                                   "Free Sulfur Dioxide","Total Sulfur Dioxide","Density","pH","Sulphates","Alcohol","Quality"],
                            skiprows=1,sep=r'\s*;\s*',engine='python')
```

```python
sns.lmplot(x="Residual Sugar",y="Density",data=white_wine_data,height=7,aspect=2)

# confidence internal is narrow here as the point  are close to the regression line
```

```python
sns.lmplot(x="Quality",y="Alcohol",data=white_wine_data)
```

```python
sns.lmplot(x="Quality",y="Alcohol",data=white_wine_data,x_jitter=.1)
```

```python
import numpy as np
sns.lmplot(x="Quality",y="Alcohol",data=white_wine_data,x_estimator=np.mean)
```
