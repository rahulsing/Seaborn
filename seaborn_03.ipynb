# plot bivariant relationship between every pair of columns in a dataset

```python
import os 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
white_wine_data=pd.read_csv("./datasets/wine-reviews/winequality-white.csv",
                            names=["Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar","Chlorides",
                                   "Free Sulfur Dioxide","Total Sulfur Dioxide","Density","pH","Sulphates","Alcohol","Quality"],
                            skiprows=1,sep=r'\s*;\s*',engine='python')
```

```python
sns.pairplot(white_wine_data,size=3)

# Diagnal is a univariant histogram
```

```python
sns.pairplot(white_wine_data,
             vars=["Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar"])
#pairwise relationship between few columns/variable
```

```python
sns.pairplot(white_wine_data,
             vars=["Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar"],
            diag_kind="kde")
#pairwise relationship between few columns/variable with diagonal as kde
```

```python
sns.pairplot(white_wine_data,
             vars=["Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar"],
            kind="reg")
#pairwise relationship between columns/variable with reggretation line fitter for each relation
```

```python
sns.pairplot(white_wine_data,
             vars=["Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar"],hue="pH")
#pairwise relationship between columns/variable with a 3rd dimention represented as Hue. The pH values of wine is represneted as Hue
```

# heat map
Find corelation between every pair of column

```python
corr_mat=white_wine_data.corr()
f, ax=plt.subplots(figsize=(10,10))
sns.heatmap(corr_mat,vmax=.8,square=True,annot=True,fmt='.2f',cmap="summer")
plt.show()

# corellation matrix is passed to heatmap function of seaborn
# square=True,annot=True : repesents the cells to be square and annotated
# along the diagonal the feature will be perfectly co-related because correlate with itself

# few more highly correlated features are like density highly correlated with Residual sugar

# this heatmap help in identifing the dimentationality reduction
```
